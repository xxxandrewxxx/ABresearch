```python
# Импорт библиотек
import numpy as np
from scipy.stats import beta, bernoulli

# Исходные данные  
a = 1
b = 1
c = 1  
d = 1   
e = 1
f = 1
epsilon = 0.1

# Инициализация
conversions_A = 0
visits_A = 0
conversions_B = 0 
visits_B = 0
conversions_C = 0
visits_C = 0
  
a_post = a
b_post = b
c_post = c
d_post = d
e_post = e
f_post = f

test_finished = 0

EL_A = 0
EL_B = 0
EL_C = 0

page_number_with_min_EL = ''

# Цикл по посетителям
for i in range(10000):

  # Выбор группы
  group = np.random.choice([1,2,3], p=[1/3, 1/3, 1/3])  

  # Моделирование поведения
  if group == 1:
    conversion = bernoulli.rvs(beta.rvs(a_post, b_post))
    conversions_A += conversion
    visits_A += 1

  if group == 2:
    conversion = bernoulli.rvs(beta.rvs(c_post, d_post))
    conversions_B += conversion
    visits_B += 1

  if group == 3:
    conversion = bernoulli.rvs(beta.rvs(e_post, f_post))  
    conversions_C += conversion
    visits_C += 1

  # Обновление параметров
  a_post = a + conversions_A
  b_post = b + visits_A - conversions_A
  c_post = c + conversions_B 
  d_post = d + visits_B - conversions_B
  e_post = e + conversions_C
  f_post = f + visits_C - conversions_C

  # Расчет ожидаемых потерь
  EL_A = (beta.mean(a_post, b_post) - 
          beta.mean(c_post, d_post) *  
          beta.mean(e_post, f_post))
   
  EL_B = (beta.mean(c_post, d_post) -
          beta.mean(a_post, b_post) *
          beta.mean(e_post, f_post))

  EL_C = (beta.mean(e_post, f_post) - 
          beta.mean(a_post, b_post) *
          beta.mean(c_post, d_post))

  # Проверка условия остановки
  if abs(min(EL_A, EL_B, EL_C) - max(EL_A, EL_B, EL_C)) > epsilon:  
    test_finished = 1
    break

# Определение лучшего варианта
if EL_A < EL_B and EL_A < EL_C:
  page_number_with_min_EL = 'A' 
elif EL_B < EL_C:
  page_number_with_min_EL = 'B'
else:
  page_number_with_min_EL = 'C'
  
# Вывод результата  
print('Оптимальный вариант:', page_number_with_min_EL)
```
